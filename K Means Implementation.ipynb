```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
plt.style.use('dark_background')
from cluster import cluster

class KMeans(cluster):
    
    def __init__(self, K=5, max_iterations=100):
        self.K = K
        self.max_iterations = max_iterations
    
    def create_random_centroids(self, X, num_columns, num_rows):

        centroids = np.zeros((self.K, num_columns))
        for k in range(self.K):
            centroid = X[np.random.choice(range(num_rows))]
            centroids[k] = centroid
        return centroids
    
    def create_cluster(self, X, centroids):
        clusters = [[] for _ in range(self.K)]
        for point_idx, point in enumerate(X):
            closest_centroid = np.argmin(
                np.sqrt(np.sum((point-centroids)**2, axis=1))
            )#Calculate Eucledian Distance
            clusters[closest_centroid].append(point_idx)
        return clusters 
    
    def new_centroids(self, cluster, X, num_columns):
        centroids = np.zeros((self.K, num_columns))
        for idx, cluster in enumerate(cluster):
            new_centroid = np.mean(X[cluster], axis=0)
            centroids[idx] = new_centroid
        return centroids
    
    def identify_cluster(self, clusters, X, num_rows):
        y_pred = np.zeros(num_rows)
        for cluster_idx, cluster in enumerate(clusters):
            for sample_idx in cluster:
                y_pred[sample_idx] = cluster_idx
        return y_pred
     
    def fit(self, X):
        num_rows, num_columns = X.shape
        centroids = self.create_random_centroids(X, num_columns, num_rows)
        for _ in range(self.max_iterations):
            clusters = self.create_cluster(X, centroids)
            prev_centroids = centroids
            centroids = self.new_centroids(clusters, X, num_columns)
            difference = centroids - prev_centroids
            if not difference.any():
                break
        y_pred = self.identify_cluster(clusters, X, num_rows) # predict function
        
        return y_pred, centroids
```

```python
kmeans = KMeans(K=4, max_iterations=100)
y_pred = kmeans.fit(X)
y_pred
```

Extend K Means

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
plt.style.use('dark_background')
from cluster import cluster

class KMeans(cluster):
    
    def __init__(self, K=5, max_iterations=100, balanced = False):
        self.K = K
        self.max_iterations = max_iterations
        self.balanced = balanced
    
    def create_random_centroids(self, X, num_columns, num_rows):

        centroids = np.zeros((self.K, num_columns))
        for k in range(self.K):
            centroid = X[np.random.choice(range(num_rows))]
            centroids[k] = centroid
        return centroids
    
    def create_cluster(self, X, centroids):
        
        if self.balanced:
            num_points = X.shape[0]
            distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
            closest_centroids = np.argmin(distances, axis=0)
            order = np.argsort(distances[closest_centroids, range(num_points)])
            clusters = [[] for _ in range(self.K)]
            for i, idx in enumerate(order):
                cluster_idx = i % self.K
                clusters[cluster_idx].append(idx)
        else:
            clusters = [[] for _ in range(self.K)]
            for point_idx, point in enumerate(X):
                closest_centroid = np.argmin(
                    np.sqrt(np.sum((point-centroids)**2, axis=1))
                )#Calculate Eucledian Distance
                clusters[closest_centroid].append(point_idx)
        
        return clusters
    
    def new_centroids(self, cluster, X, num_columns):
        centroids = np.zeros((self.K, num_columns))
        for idx, cluster in enumerate(cluster):
            new_centroid = np.mean(X[cluster], axis=0)
            centroids[idx] = new_centroid
        return centroids
    
    def identify_cluster(self, clusters, X, num_rows):
        y_pred = np.zeros(num_rows)
        for cluster_idx, cluster in enumerate(clusters):
            for sample_idx in cluster:
                y_pred[sample_idx] = cluster_idx
        return y_pred
     
    def fit(self, X):
        num_rows, num_columns = X.shape
        centroids = self.create_random_centroids(X, num_columns, num_rows)
        for _ in range(self.max_iterations):
            clusters = self.create_cluster(X, centroids)
            prev_centroids = centroids
            centroids = self.new_centroids(clusters, X, num_columns)
            difference = centroids - prev_centroids
            if not difference.any():
                break
        y_pred = self.identify_cluster(clusters, X, num_rows) # predict function
        
        return y_pred, centroids
```

```python
kmeans = KMeans(K=4, max_iterations=100, balanced = True)
y_pred = kmeans.fit(X)
y_pred
```
